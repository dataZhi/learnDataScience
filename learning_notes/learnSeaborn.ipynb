```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.randn(1000)
plt.hist(x, bins=20)
```

```python
styles = [None, 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
fig, axes = plt.subplots(2, 3, sharex=True, sharey=True)
for idx, style in enumerate(styles):
    if style:
        sns.set(style=style)
        axes.flat[idx].hist(x, bins=20)
    else:
        axes.flat[idx].hist(x, bins=20)
# fig.show()
```

```python
styles = [None, 'darkgrid', 'whitegrid', 'ticks', 'dark', 'white']
plt.subplots(2, 3, sharex=True, sharey=True, figsize=(10, 7))
for idx, style in enumerate(styles):
    if not style:
        plt.subplot(231)
        plt.hist(x, bins=20)
        plt.title('style=matplotlib')
    else:
        with sns.axes_style(style=style):
            plt.subplot(2, 3, idx+1)
            plt.hist(x, bins=20)
            plt.title('style={}'.format(style))
# plt.
```

```python
sns.set()
contexts = ['notebook', 'paper', 'talk', 'poster']
plt.subplots(2, 2, sharex=True, sharey=True, figsize=(12, 10))
for idx, context in enumerate(contexts):
    with sns.plotting_context(context=context):
        plt.subplot(2, 2, idx+1)
        plt.hist(x, bins=20)
        plt.title('context={}'.format(context))
```
